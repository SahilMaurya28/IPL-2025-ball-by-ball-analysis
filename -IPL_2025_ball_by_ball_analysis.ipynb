```python
#Import Required Libraries
```

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read dataset
```

```python
data = pd.read_csv("ipl_2025_deliveries.csv")
print(data)
```

```python
#Creating Dataframes
```

```python
df = pd.DataFrame(data)
print(df.head(5))
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#Converting columns to their correct types
```

```python
df["date"] = pd.to_datetime(df["date"])
df["date"].dt.day_name()
```

```python
#Handling missing values
```

```python
df["wicket_type"] = df["wicket_type"].fillna("no wicket")
df["player_dismissed"] = df["player_dismissed"].fillna("not out")
df["fielder"] = df["fielder"].fillna("no fielder")
```

```python
df.isna().sum()
```

```python
#Match And Season Level Insights
```

```python
#Average runs Scored in Each phase of the Game
```

```python
def phase(over):
    if 1<= over <= 6:
        return "Powerplay"
    elif 7<= over <= 15:
        return "Middle"
    else:
        return "Death"

df["phase"] = df["over"].apply(phase)
```

```python
phase_runs = df.groupby(["match_id","phase"])["runs_of_bat"].sum().reset_index()
print(phase_runs)
```

```python
avg_runs_phase = df.groupby("phase")["runs_of_bat"].mean().reset_index()
colors = ["blue","green","yellow"]
plt.bar(avg_runs_phase["phase"],avg_runs_phase["runs_of_bat"],edgecolor = "black",color = colors)
plt.xlabel("Phase of Game", fontsize = 15)
plt.ylabel("Average runs",fontsize = 15)
plt.title("Average runs Scored in different phases of the game",fontsize = 20)
plt.show()
```

```python
# Boundary Frequency
```

```python
# marking boundary balls
def is_boundary(x):
    return 1 if x in [4,6] else 0
df["is_boundary"] = df["runs_of_bat"].apply(is_boundary)
```

```python
boundary_freq = df.groupby("phase").agg(
    total_balls = ("over","count"),
    total_boundaries = ("is_boundary","sum")
).reset_index()

boundary_freq["boundary_freq%"] = boundary_freq["total_boundaries"] / boundary_freq["total_balls"] * 100
print(boundary_freq)
```

```python
colors = ["blue","green","yellow"]
plt.bar(boundary_freq["phase"],boundary_freq["boundary_freq%"],color = colors)
plt.xlabel("Phases of the game",fontsize = 15)
plt.ylabel("Percentage of boundaries",fontsize = 15)
plt.title("Boundary Frequency",fontsize = 25)
plt.show()
```

```python
#Average wickets in different phases of the game
```

```python
df["is_wicket"] = df["wicket_type"].apply(lambda x: 0 if x == "no wicket" else 1)

wickets_per_phase = df.groupby(["match_id","phase"])["is_wicket"].sum().reset_index(name="wickets")
print(wickets_per_phase)
```

```python
avg_wickets_phase = wickets_per_phase.groupby("phase")["wickets"].mean().reset_index()
plt.bar(avg_wickets_phase["phase"],avg_wickets_phase["wickets"],color = colors)
plt.xlabel("Phases of the game",fontsize = 15)
plt.ylabel("Average wickets",fontsize = 15)
plt.title("Average wickets in different phases of game",fontsize = 20)
plt.show()
```

```python
#dot ball = no runs scored od the bat
df["is_dot"] = df["runs_of_bat"].apply(lambda x: 1 if x == 0 else 0)
print(df)
```

```python
dot_balls_phase = df.groupby("phase")["is_dot"].mean().reset_index()

dot_balls_phase["dot_ball_percentage"] = dot_balls_phase["is_dot"]*100
print(dot_balls_phase)
```

```python
colors = ["blue", "green", "red"]

plt.bar(dot_balls_phase["phase"], dot_balls_phase["dot_ball_percentage"], color=colors)
plt.ylabel("Dot Ball %")
plt.title("Dot Ball Frequency per Phase")
plt.show()
```

```python
print(df.columns.tolist())
```

```python
batting_phase = df.groupby("phase").agg(
    runs_scored = ("runs_of_bat","sum"),
    balls_faced = ("runs_of_bat","count")
).reset_index()
```

```python
batting_phase["Strike_rate"] = batting_phase["runs_scored"]/batting_phase["balls_faced"]*100
print(batting_phase)
```

```python
colors = ["blue","green","yellow"]
plt.bar(batting_phase["phase"],batting_phase["Strike_rate"],color = colors)
plt.xlabel("Phase of game",fontsize = 15)
plt.ylabel("Strike rate",fontsize = 15)
plt.title("Strike rate in different phases of game",fontsize = 20)
plt.show()
```

```python
#Bowling Phase insights
```

```python
df["runs_conceded"] = df["runs_of_bat"] + df["extras"]

bowling_phase = df.groupby("phase").agg(
    runs_conceded = ("runs_conceded", "sum"),
    balls_bowled = ("runs_of_bat","count")                #counts only legal deliveries
).reset_index()

#Economy rate = Runs Conceded + Overs Bowled

bowling_phase["overs"] = bowling_phase["balls_bowled"]/6
bowling_phase["Economy_rate"] = bowling_phase["runs_conceded"]/bowling_phase["overs"]

print(bowling_phase)
plt.bar(bowling_phase["phase"],bowling_phase["Economy_rate"],color = colors)
plt.xlabel("phase of game",fontsize = 15)
plt.ylabel("Economy rate",fontsize = 15)
plt.title("Economy rate of bowlers in different phases of the game",fontsize = 20)
plt.show()
```

```python
# Wickets per over
wickets_over = df[df["is_wicket"]==1].groupby(["match_id","over"]).size().reset_index(name="wickets")

# Highlight overs with 2+ wickets
for _, row in wickets_over.iterrows():
    if row["wickets"] >= 2:
        print(f"Turning Point: In match {row['match_id']}, over {row['over']} saw {row['wickets']} wickets fall!")
```

```python
# Boundaries per over
boundaries_over = df[df["is_boundary"]==1].groupby(["match_id","over"]).size().reset_index(name="boundaries")

# Highlight overs with 2+ boundaries
for _, row in boundaries_over.iterrows():
    if row["boundaries"] >= 2:
        print(f"Turning Point: In match {row['match_id']}, over {row['over']} had {row['boundaries']} boundaries!")
```

```python
# Runs per over
runs_over = df.groupby(["match_id","over"])["runs_of_bat"].sum().reset_index()

# Look for overs with 10+ runs
for _, row in runs_over.iterrows():
    if row["runs_of_bat"] >= 10:
        print(f"Turning Point: In match {row['match_id']}, over {row['over']} exploded for {row['runs_of_bat']} runs!")
```
